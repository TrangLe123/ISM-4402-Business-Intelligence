```python
import pandas as pd
Location = "C:/Users/Trang Le/OneDrive/Desktop/datasets/axisdata.csv"
df = pd.read_csv(Location)
df.head(100)
```

```python
#1 Average Cars Sold per month
df['Cars Sold'].mean()
```

```python
#2 Max Cars Sold per month
df['Cars Sold'].max()
```

```python
#3 Min Cars Sold per month
df['Cars Sold'].min()
```

```python
#4 Average Cars Sold per month by gender
pd.pivot_table(df, values = ['Cars Sold'], index = ['Gender'], aggfunc = 'mean')
```

```python
#5 Average Hours Worked by people selling more than 3 cars per month
df1 = df.loc[df['Cars Sold'] > 3]
pd.pivot_table(df1, index = ['Fname'], values = ['Hours Worked'], aggfunc = 'mean')
```

```python
#6 Average Years of Experience
df['Years Experience'].mean()
```

```python
#7 Average Years of Experience for people selling more than 3 cars per month
df2 = df.loc[df['Cars Sold'] > 3]
pd.pivot_table(df2, index = ['Fname'], values = ['Years Experience'], aggfunc = 'mean')
```

```python
#8 Average Cars Sold per month sorted by whether they have had sales training.
df3 = df.loc[df['SalesTraining'] == 'Y']
pd.pivot_table(df3, index = ['Fname', 'Lname'], values = ['Cars Sold'], aggfunc = 'mean')
```
