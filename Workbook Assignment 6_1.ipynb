```python
import pandas as pd
Location = "C:/Users/Trang Le/OneDrive/Desktop/datasets/gradedata.csv"
df = pd.read_csv(Location)
df.head()
```

```python
def score_to_numeric(x):
    if x=='female':
        return 1
    if x=='male':
        return 0
df['GenderinNumericValue'] = df['gender'].apply(score_to_numeric) 
df.tail()
```

```python
import statsmodels.formula.api as sm
result = sm.ols(
 formula='grade ~ exercise + hours + GenderinNumericValue',
 data=df).fit()
result.summary()
```

```python
#Yes, R-squared did improve from 0.664 to 0.665 after adding the numeric value of gender. 
```
