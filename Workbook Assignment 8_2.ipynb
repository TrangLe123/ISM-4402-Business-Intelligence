```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
Location = "C:/Users/Trang Le/OneDrive/Desktop/datasets/gradedata.csv"
df = pd.read_csv(Location)
df.plot.scatter(x = 'hours', y = 'grade')
plt.xlabel('Hours')
plt.ylabel('Grade')
plt.show()
```

```python
#Explanation: Hours (X) is an independent variable, Grade (Y) is a dependent variable. Because I have only one independent
# variable so it is the simple linear regression. As can be seen in the chart, I can see that the trendline of the scatter
# plot is the positive linear trajectory. I used the MS Excel to calculate the correlation between these two variables and 
# I got r = 0.8, so it has a pretty strong correlation. However, the R-squared is only 0.64 (64%) by using MS Data Analysis
# which is quite low. I could say 64% of the total variation in the number of grades is explained by the number of hours
# that students study. Thus, the more hour for studying, the higher grades students can earn. 
#In the scatter plot, I can see that a couple special cases that not too many studying hours, they still can earn high 
# grades, and one case that spending a lot hours for studying but the grade is still low. I believe that is why R-squared 
# has only 64% (low).
```
